```r
library(dplyr)
```

```r
#Filter all male left handers.
filter(new_survey,W.Hnd == 'Left')
```

```r
#Display all female right handers who keep right on top while clapping.
filter(new_survey,Sex == 'Female', W.Hnd == 'Right' , Clap == 'Right' )
```

```r
#Display all students who never exercise.
filter(new_survey,Exer == 'None' )
```

```r
#Display only the gender, age and writing hand of the students.
head(select(new_survey,Sex,Age,W.Hnd))
```

```r
#Display the age, pulse rate and writing hand span of female left handers.
new_survey %>%
filter(Sex == 'Female',W.Hnd == 'Left') %>%
select(Age,Pulse,Wr.Hnd)
```

```r
#Arrange all male left handers according to descending order of their heights.
new_survey %>%
filter(Sex == 'Male',W.Hnd == 'Left') %>%
arrange(desc(Height))

```

```r
#Introduce a new column hand_span which contains the value as difference between the span of writing hand and
#non-writing hand and 
#display it along with gender, writing hand and non-writing hand span.
new_survey %>%
mutate(hand_span = Wr.Hnd - NW.Hnd)%>%
select(Sex,Wr.Hnd,NW.Hnd,hand_span)
```
